```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from scipy.special import softmax,expit
```

```python
nsamples_per_class = 500
dim = 2
hidden_layer = 3
no_of_classes = 3
X1 = np.random.randn(nsamples_per_class,dim) + np.array([0,-2])
X2 = np.random.randn(nsamples_per_class,dim) + np.array([2,2])
X3 = np.random.randn(nsamples_per_class,dim) + np.array([-2,2])
X = np.vstack([X1,X2,X3])
Y = np.array([0]*nsamples_per_class+ [1]*nsamples_per_class + [2]*nsamples_per_class)
fig = go.Figure()
fig = px.scatter(x=X[:,0],y=X[:,1],color=Y)
fig.show()
```

```python
### Exerrcise 
X = np.array([1,2])
W1 = np.array([[1,1],[1,0]])

W2 = np.array([[0,1],[1,1]])

a1 = X.dot(W1)
a2 = np.tanh(a1)
output = softmax(a2.dot(W2),axis=0)
print(output)
```

```python
W1 = np.random.randn(dim,hidden_layer)
b1 = np.random.randn(hidden_layer)

W2 = np.random.randn(hidden_layer,no_of_classes)
b2 = np.random.randn(no_of_classes)

a1 = X.dot(W1) + b1
a2 = expit(a1)
output = softmax(a2.dot(W2) + b2,axis=1)
print(output[0])
```
