```python
import pandas as pd
import numpy as np
from plotly.offline import init_notebook_mode
import plotly.graph_objects as go
init_notebook_mode(connected=True)
```

```python
t = 1
y = np.linspace(0.00001,0.999,10000)
J_1 = -(t * np.log(y) + (1 - t)*np.log(1-y))


t = 0
y = np.linspace(0.00001,0.999,10000)
J_2 = -(t * np.log(y) + (1 - t)*np.log(1-y))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=y,y=J_1,mode ='lines'))
fig.add_trace(go.Scatter(x=y,y=J_2,mode ='lines'))
fig.show()
```

```python
def univariate_normal(x,mu,sigma):
    power_term = ((x - mu)/sigma)**2
    normal_term = (np.sqrt(2*np.pi*sigma**2))
    return np.exp(-1/2*power_term)/normal_term
    
```

```python
mu = 100
sigma = 100
x = np.linspace(0,200,1000)
y=univariate_normal(x,mu,sigma)
s = np.random.normal(mu, sigma, 10000)
fig = go.Figure()
fig.add_trace(go.Scatter(x=x,y=y,mode ='lines'))
fig.add_trace(go.Histogram(x=s,histnorm='probability density'))
fig.show()
```
