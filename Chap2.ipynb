```python
# All bokeh library
from bokeh.io import output_notebook
import bokeh
from bokeh.plotting import figure, show
from bokeh.charts import Histogram
from bokeh.palettes import Viridis256
print(bokeh.__version__)
TOOLS="hover"
output_notebook()
from sklearn.datasets import load_iris
import numpy as np
import scipy as sp
from bokeh.sampledata.iris import flowers as data_iris
import itertools
from bokeh.layouts import gridplot
output_notebook()
import pandas as pd
```

```python
data_iris.head()
```

```python
list_headers = list(data_iris.columns.values) 
del list_headers[-1]
```

```python
list_combination = list(itertools.combinations(list_headers,2))
```

```python
p = figure(tools=TOOLS)
```

```python
data_iris.species.unique()
```

```python
# data_iris[[list_combination[0][0],list_combination[0][1]]][data_iris["species"] == 'setosa']
# list_combination[0][0],list_combination[0][1]
def iris_scatter_plot(list_combination):
    list_figures =[]
    for index in list_combination:
        p = figure(tools=TOOLS)
        data_setosa = data_iris[[index[0],index[1]]][data_iris["species"] == 'setosa']
        data_versicolor = data_iris[[index[0],index[1]]][data_iris["species"] == 'versicolor']
        data_virginica = data_iris[[index[0],index[1]]][data_iris["species"] == 'virginica']        
        p.triangle(data_setosa.iloc[:,0],data_setosa.iloc[:,1],color = 'brown',legend = 'setosa')
        p.circle(data_versicolor.iloc[:,0],data_versicolor.iloc[:,1],color = 'green',legend = 'versicolor')
        p.diamond(data_virginica.iloc[:,0],data_virginica.iloc[:,1],color = 'navy',legend = 'virginica')
        p.xaxis.axis_label = index[0]
        p.yaxis.axis_label = index[1]
        p.legend.location = "top_left"
        p.legend.orientation = "horizontal"
        list_figures.append(p)
    grid = gridplot(list_figures, ncols=3, plot_width=300, plot_height=300)
    show(grid)
```

```python
iris_scatter_plot(list_combination)
```
