```python
import requests
from bs4 import BeautifulSoup
from collections import OrderedDict
import json
import pickle

r = requests.get("http://data-interview.enigmalabs.org/companies/")

dict_json= {}
list_elem =[]

dict_json['company_listing'] = list_elem

soup = BeautifulSoup(r.content)
#soup.contents
#for a in soup.find_all('a', href=True):
    #print "Found the URL:", a['href']
table = soup.find('table', 'table table-hover')

for a in table.findAll('a',href=True):
     #print a['href']
     url = "http://data-interview.enigmalabs.org"+ a['href']
     r = requests.get(url)
     soup= BeautifulSoup(r.content)
     table_data = [[cell.text for cell in row("td")] for row in soup("tr")]
     json_str= OrderedDict(table_data) 
     list_elem.append(json_str)

output = open('data.json', 'w+')

json.dump(dict_json,output)
output.close()
dict_json
```
